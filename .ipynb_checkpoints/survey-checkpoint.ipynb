```python
#visualize data with matplotlib
from matplotlib import pyplot as plt
percent_spread = (data['EdLevel'].value_counts(normalize=True) / 89184) * 100 #Series
percent_spread = percent_spread.sort_values()
percent_spread.plot(kind='barh')
plt.title('Distribution of education levels')
plt.xlabel('Percentage')
plt.ylabel('Educaton level')
plt.show()
```

```python
#calculating offline learning percentage share
offline_learning = data['LearnCode'].str.split(";").explode() #list içindeki verileri ayrı rowlara dönüştürür
method_counts = offline_learning.value_counts(normalize = True) * 100
method_counts = method_counts.sort_values()
method_counts.plot(kind='barh')
plt.title('Distribution of offline learning methods')
plt.xlabel('Percentage')
plt.ylabel('Offline Learning Methods')
for index,value in enumerate(method_counts): #index ile beraber
    plt.text(value, index, f'{value:.2f}%')
plt.show()
```

```python
#Platforms percentage share
import numpy as np
online_platform_responses = data[data['LearnCodeCoursesCert'] != np.nan]
response_count = online_platform_responses['LearnCodeCoursesCert'].count()

online_platform = data['LearnCodeCoursesCert'].str.split(";").explode()
online_platform_counts = online_platform.value_counts()
online_platform_percentages = (online_platform_counts/response_count) * 100

online_platform_percentages = online_platform_percentages.sort_values()
online_platform_percentages.plot(kind="barh")
plt.title('Online plartforms')
plt.xlabel('Percentage')
plt.ylabel('Learn Code Online Platforms')
for index,value in enumerate(online_platform_percentages): #index ile beraber
    plt.text(value, index, f'{value:.2f}%')
plt.show()
```
